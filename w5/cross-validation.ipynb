```python
def get_best_weight(x, y, reg):

    dims = x.shape[0]
    reg_diag = np.zeros( (dims,dims) )
    np.fill_diagonal( reg_diag, reg )
    
    first_term = np.dot( x, x.T ) + reg_diag
    inv_first = np.linalg.inv(first_term)
    second_term = np.dot(x,y)
    
    return np.dot(inv_first, second_term)

lambda_basket = np.arange(-4,4.1,0.1)
reg_basket = np.power(10, lambda_basket) # from [-4,4]

# weight_basket = []
# for reg in reg_basket:
#     w = get_best_weight(x_expanded, training_set.obs, reg )
#     weight_basket.append(w)

# print(weight_basket[0].shape)
```

```python
no_fold = 10

# for train_index, test_index in kf:
#     print(len(train_index))
#     print(len(test_index))
#     print('----')
# train_idx = np.arange( 0, x_expanded.shape[1] )
# k_folds = np.split(train_idx,no_fold)
# print(train_idx)
# print(k_folds)
# print(t)

def predict(x, w):
    return np.dot(w.T,x)

def mse(pred, actual):
    n = pred.shape[0]
    delta = pred - actual
    return np.dot( delta.T, delta ) / n 

def train_with_kfold(x, y,  reg, kf ):
    error_basket = []
    best_w = None
    best_err = math.inf
    for train_index, test_index in kf:
        w = get_best_weight( x[:, train_index], y[train_index], reg )
        preds = predict(x[:,test_index], w)

        err = mse(preds, y[test_index])
        if err < best_err:
            best_err = err
            best_w = w
        error_basket.append(err)
        
    return  { 'best_w': best_w, 'best_err': best_err, 'mean_mse': np.mean(error_basket), 'var_mse': np.var(error_basket), 'raw_errors': error_basket }


def train_with_regs(x,y, reg_values, no_fold=10):
    results = []
    
    kf = KFold(x.shape[1], no_fold)
    for reg in reg_basket:
        res = train_with_kfold(x,y, reg, kf )
        results.append(res)
    return results

train_result_from_regs_df = pandas.DataFrame(train_with_regs( x_expanded, training_set.obs, reg_basket ))
```

```python
# RAW_ERROR = train_result_from_regs.raw_errors.as_matrix()
plt.rcParams["figure.figsize"] = [10,3]
def plot_mse_stats( cv_result_df ):
    p = plt.subplot(1,2,1)
    p.plot(lambda_basket, cv_result_df.mean_mse)

    p.set_title("Mean of MSE per CV")
    p.set_xlabel("lambda = 10^{x}")
    
    p = plt.subplot(1,2,2)
    p.plot(lambda_basket, cv_result_df.var_mse)

    p.set_title("Variance of MSE per CV")
    p.set_xlabel("lambda = 10^{x}")
    plt.tight_layout()

plot_mse_stats( train_result_from_regs_df )
```

```python
best_lambda_idx = np.argmin(train_result_from_regs_df.mean_mse)
best_lambda = reg_basket[best_lambda_idx]
best_w = train_result_from_regs_df.loc[best_lambda_idx].best_w



def plot_predicted_values_validation_set( x, original_x, actual_y, w):
    
    
    
    pred_val = predict(x, w)
    
    vmin = np.min(actual_y)
    vmax = np.max(actual_y)
    
    fig, ax = plt.subplots( 1, 2, sharex=True, sharey=True)

    ax = ax.flatten()
    intensities = [pred_val, actual_y]
    labels = ["prediction","actual"]
    for i in range(len(ax)):
        plt.subplot(1,2,i+1)
        plt.scatter( original_x[0,:], original_x[1,:], c=intensities[i], s=15, edgecolor='w', vmin=vmin, vmax=vmax)
        plt.colorbar()
        plt.ylabel('x2')
        plt.xlabel('x1')
    
        plt.title(labels[i])
        plt.xlim(135,215)
        plt.ylim(35,125)



    plt.tight_layout()

    
plot_predicted_values_validation_set( val_x_expanded, val_x, validation_set.dens, best_w )
```
