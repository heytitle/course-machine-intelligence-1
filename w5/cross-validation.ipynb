```python
def find_transformation_matrix(x):
    n = x.shape[1]
    cov_matrix = (1.0/n)*( np.dot(x, x.T))
    (eig_vals, eig_vecs) = np.linalg.eig(cov_matrix)

    rotation_matrix = eig_vecs.T
    scaling_matrix = np.diag(1/(np.sqrt(eig_vals)))
  
    trans_matrix = np.dot( scaling_matrix, rotation_matrix )

    return ( transform(trans_matrix,x), trans_matrix )
#     return (np.dot(trans_matrix, x), trans_matrix)

def transform( trans_matrix, x ):
    return np.dot(trans_matrix, x )

(x_whitened, trans_matrix ) = find_transformation_matrix(x_center)

plot_scatter(x_whitened, title="Data after whitening")

print("mean after whitening")
print(np.mean( x_whitened, axis=1)) # axis = 1 : calculate mean by row

print("variance after whitening")
print(np.var( x_whitened, axis=1))
```

```python
def expand_dimension(x, k):
    new_dims = int((k+1)*(k+2)/2)
    n = x.shape[1]
    new_x = np.zeros( (new_dims, n ) )
    row_index = 0
    for d in range(k+1):
#         print("d" + str(d))
        for i in range(d+1):
            new_x[row_index,:] = np.multiply(np.power(x[0,:],i), np.power(x[1,:], d-i))
            row_index = row_index + 1
#     print(new_x)
#     print(new_x.shape)
    return new_x
    
# toy = np.matrix( [ [1,3], [2,4] ])
# expand_dimension(toy,2)
x_expanded = expand_dimension( x_whitened, 9)

plt.pcolor( x_expanded[3,:].reshape(20,10), cmap=plt.cm.Blues)
plt.title('dimension 3 x1^0, x2^2 for training data')
# heatmap = ax.pcolor(heat_accs, cmap=plt.cm.Reds)
```
